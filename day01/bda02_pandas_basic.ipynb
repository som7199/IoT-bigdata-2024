#### 판다스 기본 
- 시리즈, 데이터프레임 개념
- 판다스에서는 데이터프레임이 가장 기본, 데이터 처리 도중 잘라서 시리즈 사용

<img src="https://raw.githubusercontent.com/hugoMGSung/bigdata-analysis-2024/main/images/ba001.png" width="730">

````python
# 데이터프라임 객체 - 2차원 배열
## 1. 딕셔너리로 DataFrame 생성
origin = {'이름' : ['김소민', '이경민', '꽉철이'], '나이' : [24, 25, 5], '결혼' : [True, True, False]}
origin, type(origin)
````

````python
## 딕셔너리로 데이터프레임 생성, 변수명을 보통 df로 사용
df = pd.DataFrame(data=origin)
df
````

````python
# 데이터 프레임 타입
type(df)
````

````python
## 2. 파이썬 리스트로 데이터 프레임 생성
data = [['김소민', 24, '김해', True],
        ['이경민', 25, '부산', True],
        ['꽉철이', 5, '서울', False],
        ['솜아지', 4, '수원', False],]

df2 = pd.DataFrame(data=data)
df2
````

````python
## 리스트로 DF 생성 시 컬럼 제목이 없으므로 추가해줘야함!
df2 = pd.DataFrame(data=data, columns=['이름', '나이', '주소', '결혼'])
df2
````

````python
## 시리즈 생성
sr1 = pd.Series([1, 2, 3, 4])
sr1
````

````python
# 시리즈 타입 확인
type(sr1)
````
